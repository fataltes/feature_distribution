```python
db = 'twitterGender'
import sqlalchemy
from sqlalchemy import create_engine
from sqlalchemy.engine.url import URL
myDB = URL(drivername='mysql', database=db, query={ 'read_default_file' : '/home/fatal/.my.cnf' })
engine = create_engine(name_or_url=myDB, encoding='utf8')
#conn = engine.connect()
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

plt.figure()
usr_freq = pd.read_sql('select * from msgs_usr_cnt', con=engine)
plt.plot(usr_freq.cnt, usr_freq.freq)
print("user message frequency out of {}".format(usr_freq.freq.sum()))
print(usr_freq)

print('\n\n')

plt.figure()
cnty_freq = pd.read_sql('select * from msgs_cnty_cnt', con=engine)
plt.plot(cnty_freq.cnt, cnty_freq.freq)
print("county message frequency out of {}".format(cnty_freq.freq.sum()))
print(cnty_freq)
```
